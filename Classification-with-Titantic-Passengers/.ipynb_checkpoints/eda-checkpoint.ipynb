### The Variables & Observations

The training data contains 891 observations with 12 variables, and the test set contains 418 observation with 11 variables. Each observation is a passenger and the target variable for this analysis is Survived.

```python
print(train_df.shape)
print(test_df.shape)
```

After combining the two data sets and using the `pandas` `info` & `head` methods, we observe the following:

- Since the PassengerID is not described in the data dictionary, it appears that this is not a variable and can be dropped from the analysis. It also looks like it is just a primary record key for the table.

- Of the 10 predictor variables, four have missing values: `Age`, `Fare`, `Embarked` & `Cabin`. We will have to consider imputing these variables. 

    - For `Embark`, let's impute the 2 missing values using the mode, which is 'S' for Southampton.
    - The `Cabin` variable is more than 75% missing. Instead of dropping the variable because of the missingness, we will replace the missing values with 'Unk'.
    - For `Age` and `Fare`, let's impute using the mean value. Let's include Scikit's `Imputer` in our data pipeline to do so. 

- The data set contains 6 categorical variables. The variables `Name`, `Sex`, `Ticket`, `Cabin`, `Embarked` are class `objects`, and `Pclass` is a categorical variable that was encoded as numeric. We will want to remember to coerce `Pclass` to a categorical variable, so that the model does not think it is continuous.

- The remaining 4 predictor variables are numeric: `SibSp`, `Parch`, `Age` and `Fare`

```python
test_df['Survived'] = np.nan
all_data = pd.concat([train_df, 
                      test_df],
                    sort=False)
all_data.info()
all_data.head(5)
```

### Numeric Variables

Looking at the summary statistics & `scattermatrix` for the 4 numerical variables, we observe the following: 

- The mean and median passenger `Age` was nearly 30 and 28 years old, respectively.
- The `Fare` amount, `SibSp` and `Parch` are right skewed.
- Looking at the median values for `SibSp` and `Parch`, the typical passenger did not have any parents, children or siblings aboard the Titanic.

```python
numeric_vars = all_data[['Age', 'SibSp', 'Parch', 'Fare']]
(numeric_vars
    .describe()
    .round(decimals=1)
)
```

```python
pd.plotting.scatter_matrix(numeric_vars, 
                           figsize=(12, 12), 
                           diagonal='kde')
#https://pandas.pydata.org/pandas-docs/stable/visualization.html
```

### Categorical & Discrete Numeric Variables

Using the barplots and frequency tables below, let's examine the categorical variables. We will include numeric variables with a limited number of discrete values as well. We can make the following observations:

- Most passengers did not survive.

- 3rd class passengers outnumbered the 1st and 2nd class ones.

- The `Name`, `Ticket` & `Cabin` variables have more values than can be comfortably displayed in barplots.

- There were 300 males more males than females were aboard.

- The overwhelming majority embarked from Southampton


```python
categorical_vars = all_data.drop(['PassengerId', 'Age', 'Fare'], axis=1)

plt.figure(figsize=(16, 16))
for i, column in enumerate(categorical_vars.columns):
    plt.subplot(3, 3, i + 1)
    sns.countplot(x=column, data=categorical_vars) #https://seaborn.pydata.org/tutorial/categorical.html   
plt.show()
```

Finally, let's take a close look at variables `Name`, `Ticket` & `Cabin` by looking at the top 15 most frequent values. They have many unique values and will create a lot of dimensionality in our models. We may want to consider feature engineering these to be more meaningful variables or consider dropping them. 

- For `Name`, let's try using only the surname.
- For `Ticket`, let's drop the variable because I don't see any metadata that can be gleaned.
- For `Cabin`, the [Kaggle forums](https://www.kaggle.com/c/titanic/discussion/4693) offer evidence that the 1st alpha character represents the Deck where the room has located. Consequently, let's create a Deck variable and drop the room number and carry over the 'Unk' replacement value.

```python
high_dim_vars = categorical_vars[['Name', 'Ticket', 'Cabin']]

for column in high_dim_vars.columns:
    print('%s has %s unique values:\n' % (column, high_dim_vars[column].nunique()))
    print(high_dim_vars[column].value_counts()[0:15, ], '\n')
```
